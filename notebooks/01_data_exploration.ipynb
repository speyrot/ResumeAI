```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Update the path below to match your file structure
dataset_path = '../data/raw/Resume.csv'
df = pd.read_csv(dataset_path)

# Display the first few rows of the dataset
df.head()

# Describe the dataset to get an idea of the numerical distributions
df.describe()

# Get a summary to see data types and potential missing values.
df.info()

# Check unique categories and counts
print(df['Category'].value_counts())

# Examine the distribution of job categories since it's crucial for your model's performance across different classes.
df['Category'].value_counts().plot(kind='barh', figsize=(10, 8))
plt.title('Distribution of Job Categories')
plt.xlabel('Number of Resumes')
plt.ylabel('Job Category')
plt.show()

# Explore the length of the resumes. This could be important for understanding the variability in your data and could impact feature extraction techniques.
df['Resume_length'] = df['Resume_str'].apply(len)
sns.histplot(df['Resume_length'])
plt.title('Distribution of Resume Lengths')
plt.xlabel('Length of Resume')
plt.ylabel('Frequency')
plt.show()

#Check for and visualize missing values. This will inform your data cleaning process.
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

```
